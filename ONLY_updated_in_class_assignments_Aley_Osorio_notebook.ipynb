```python
def quicksort(L, reverse=False):
    if len(L) < 2:
        return L
    else:
        pivot_idx = len(L) // 2  # choose the index of the pivot item
        pivot = L[pivot_idx]     # this is the pivot item

        less, equal, greater = [], [], []

        for item in L:
            if item < pivot:
                less.append(item)
            elif item == pivot:
                equal.append(item)
            else:
                greater.append(item)

        if not reverse:
            return quicksort(less) + equal + quicksort(greater)
        else:
            return quicksort(greater, reverse=True) + equal + quicksort(less, reverse=True)


# Example use:
print(quicksort([8, 0, 2, 6, 1, 7, 3], reverse=True))
```

#NEW IN CLASS ASSIGNMENT QUALITY CONTROL MANAGER (EXCEPTIONS)

```python
def getValues(n: int) -> list:
    '''
    Create and return a list of weights

    Inputs:
    - n: number of items in the list

    Outputs:
    - A list of weights
    '''

    items = []

    for i in range(n):
        try:
            item = float(input('Enter a weight value: '))
        except Exception as e:
            print(e)
            item = float('nan')
        items.append(item)

    return items

def calcAverage(alist: list)-> float:
  '''
  Returns the average of list alist


  Input:
  - alist: list of weights


  Output:
  - The average value of the list
  '''
  try:
      return sum(alist[2:]) / len(alist[2:])
  except Exception as e:
      print(e)
      return float('nan')


def main():
    num = int(input('How many Weights? '))
    if num <= 0:
        raise ValueError('num must be positive')


    weights = getValues(num)
    avg = calcAverage(weights)
    print(f'Average weight: {avg} lbs.')


main()
```

#Greeting Class 11-13-25

Write a Python class and populate it with two attributes and methods:
 Attributes: first/last names
Method display(): takes a Boolean input parameter; displays the attribute values either formally (last, first) or informally (first)
Special method __str__()

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def __str__(self):
    return "Instance of class Greeting"

  def display(self, reverse=True):
    if reverse:
      print(f'{self.last}, {self.first}')
    else:
      print(self.first)

name = Greeting('John', 'Doe')
print(name)
name.display()
name.display(reverse=False)
```

#Classes Point & Rectangle

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** .5

class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(4,0)
    self.p3 = Point(4,4)
    self.p4 = Point(0,4)
    self.width = self.p1.distance(self.p2)
    self.height = self.p2.distance(self.p3)

  def area(self):
    return self.width * self.height
```
